**Bloxpot**

```python
plt.figure(figsize=(8,5))
sns.boxplot(y = "engineSize", data=df, showfliers=False)
```

**Separate the Numerical Columns**

```python
num_cols = []

for col in df.columns:
    if df[col].dtypes != "O":
        num_cols.append(col)

print(num_cols)
```

```python
num_cols[4]
```

**For loop to show the boxplots of all numeric features**

```python
plt.figure(figsize=(16,8)) # width, height

for index in range(len(num_cols)):

    plt.subplot(2,3,index+1)
    sns.boxplot(x=num_cols[index], data=df, showfliers=False)

plt.suptitle("Boxplot of the numeric columns in the Ford Used Car dataset")
plt.tight_layout()
plt.show()
```

**Adding categorical features in the boxplot**

```python
plt.figure(figsize=(20,8)) # width, height

for index in range(len(num_cols)):
    plt.subplot(2,3,index+1)
    sns.boxplot(x=num_cols[index], data=df, showfliers=False, y=df["fuelType"])

plt.suptitle("Boxplot of the numeric columns in the Ford Used Car dataset")
plt.tight_layout()
plt.show()
```

```python
df["fuelType"].value_counts()
```

**Adding categorical features (hue) in the boxplot**

```python
plt.figure(figsize=(25,12)) # width, height

for index in range(len(num_cols)):
    plt.subplot(2,3,index+1)
    sns.boxplot(y=df[num_cols[index]], x=df["mileage"], hue=df["fuelType"], showfliers=False)

plt.suptitle("Boxplot of the numeric columns in the Ford Used Car dataset")
plt.tight_layout()
plt.show()
```

**Comparing boxplots and violinplot**

```python
plt.figure(figsize=(20,8)) # width, height

for index in range(len(num_cols)): 
    plt.subplot(2,3,index+1)
    sns.violinplot(x=num_cols[index], data=df, showfliers=False)

plt.suptitle("Boxplot of the numeric columns in the Ford Used Car dataset")
plt.tight_layout()
plt.show()
```

A violin plot is more informative than a plain box plot. Box plots often misrepresent distributions but violin plot gives accurate distributions of data.

**Histogram using Seaborn**

```python
plt.figure(figsize=(20,12))

plt.subplot(2,3,1)
sns.histplot(x="mileage", data=df)
plt.title("row=1, col=1, position=1")

plt.subplot(2,3,2)
sns.histplot(x="mileage", data=df, binwidth=50)
plt.title("row=1, col=2, position=2")

plt.subplot(2,3,3)
sns.histplot(x="mileage", data=df, binwidth=50)
plt.xticks(np.arange(df["mileage"].min(), df["mileage"].max()+51, 50), rotation=90)
plt.title("row=1, col=3, position=3")

plt.subplot(2,3,4)
sns.histplot(x="mileage", data=df, binwidth=25, color="green")
plt.title("row=2, col=1, position=4")

plt.subplot(2,3,5)
sns.histplot(x="mileage", data=df, binwidth=25, color="green", stat="probability")
plt.title("row=2, col=2, position=5")

plt.subplot(2,3,6)
sns.histplot(x="mileage", data=df, binwidth=25, color="green", fill=False)
plt.title("row=2, col=3, position=6")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,16))

plt.subplot(2,2,1)
sns.histplot(x="mileage", data=df, binwidth=1)
plt.title("row=1, col=1, position=1")

plt.subplot(2,2,2)
sns.histplot(x="mileage", data=df, binwidth=10, cumulative=True)
plt.title("row=1, col=2, position=2")

plt.subplot(2,2,3)
sns.histplot(x="mileage", data=df, binwidth=50, color="green")
plt.xticks(np.arange(df["mileage"].min(), df["mileage"].max()+51, 50), rotation=90)
plt.grid()
plt.yticks(np.arange(0,4000,250))
plt.title("row=2, col=1, position=3")

plt.subplot(2,2,4)
sns.histplot(x="mileage", data=df, binwidth=50, cumulative=True, color="orange")
plt.yticks(np.arange(0,4000,250))
plt.xticks(np.arange(df["mileage"].min(), df["mileage"].max()+51, 50), rotation=90)
plt.grid()
plt.title("row=2, col=2, position=4")

plt.tight_layout()
plt.show()
```
